```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
Datos=pd.read_excel('Datos_proyecto_estadistica.xlsx')
Datos.head()
```

```python
Datos.set_index('Dates', inplace=True)
Datos.head()
```

```python
desplazado=Datos['GRUPOARG CB Equity'].shift(1) #Opcion 1
Datos['Rentabilidad GRUPOARG']=Datos['GRUPOARG CB Equity']/desplazado-1
Datos.dropna(inplace=True)
desplazado=Datos['BIPC US Equity'].shift(1) #Opcion 1
Datos['Rentabilidad BIPC']=Datos['BIPC US Equity']/desplazado-1
Datos.dropna(inplace=True)
Datos.head()
```

```python
fig1 = make_subplots(rows=1, cols=2, specs=[[{"type": "scatter"}, {"type": 'scatter'}]])
fig1.add_trace(go.Scatter(x=Datos.index,y=Datos["GRUPOARG CB Equity"]), row=1, col=1)
fig1.add_trace(go.Scatter(x=Datos.index,y=Datos["BIPC US Equity"]), row=1, col=2)
fig1.show()
```

```python
fig = px.line(Datos, x=Datos.index, y=Datos.columns[2:5])
fig.show()
```

```python
fig=px.histogram(Datos,x=Datos.columns[2:5])
fig.show()
```

```python
fig=px.box(Datos,y=Datos.columns[2:5])
fig.show()
```

```python
fig1 = make_subplots(rows=1, cols=2, specs=[[{"type": "scatter"}, {"type": 'scatter'}]])
fig1.add_trace(go.Scatter(x=Datos.index,y=Datos["Rentabilidad GRUPOARG"], mode='markers'), row=1, col=1)
fig1.add_trace(go.Scatter(x=Datos.index,y=Datos["Rentabilidad BIPC"], mode='markers'), row=1, col=2)
fig1.show()
```
